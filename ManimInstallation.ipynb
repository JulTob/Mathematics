# Create Shapes

```python
class CreateCircle(Scene):
    def construct(self):
        circle = Circle()
            # create a circle object
        circle.set_fill(RED, opacity=0.5)
            # set the color and transparency
        self.play(Create(circle))
            # show the circle on screen
```

```python
%%manim -v WARNING -qm --renderer=cairo CreateCircle
```

```python
class CreateShapes(Scene):
    def construct(self):
        circle = Circle()
        # create a circle
        circle.set_fill(RED, opacity=0.5)
        # set color and transparency

        square = Square()
        # create a square
        square.rotate(PI / 4)
        # Set the rotation to a initial amount

        self.play(Create(square))
        # animate the creation of the square
        self.play(Transform(square, circle))
        # interpolate the square into the circle
        self.play(FadeOut(square))
        # fade out animation
```

```python
%%manim -v WARNING -qm --renderer=cairo CreateShapes
```

# Positioning

```python
class SquareAndCircle(Scene):
    def construct(self):
        circle = Circle()
        # create a circle
        circle.set_fill(RED, opacity=0.5)
        # set the color and transparency

        square1 = Square()
        square2 = Square()
        square3 = Square()
        square4 = Square()
        # create squares
        square1.set_fill(BLUE, opacity=0.5)
        square2.set_fill(GREEN, opacity=0.5)
        square3.set_fill(ORANGE, opacity=0.5)
        square4.set_fill(GOLD, opacity=0.5)
        # set the color and transparency

        square1.next_to(circle, UP, buff=0.5)
        square2.next_to(circle, DOWN, buff=0.8)
        square3.next_to(circle, LEFT, buff=2.5)
        square4.next_to(circle, RIGHT, buff=1.7)
        # set the position
        self.play(Create(circle),
                  Create(square1),
                  Create(square2),
                  Create(square3),
                  Create(square4)
                  )  # show the shapes on screen
```

```python
%%manim -v WARNING -qm --renderer=cairo SquareAndCircle
```

```python
class AnimatedSquareToCircle(Scene):
    def construct(self):
        circle = Circle()
        # create a circle
        square = Square()
        # create a square

        self.play(Create(square))
        # show the square on screen
        self.play(square.animate.rotate(PI / 4))
        # rotate the square
        self.play(Transform(square, circle))
        # transform the square into a circle
        self.play(
            circle.animate.set_fill(RED, opacity=0.5)
        )  # color the circle on screen
        self.play(circle.animate.flip(UP))
        # flip the circle vertically
        self.play(circle.animate.shift(RIGHT * 2))
        # shift the circle to the right
```

```python
%%manim -v WARNING -qm --renderer=cairo AnimatedSquareToCircle
```

```python
class DifferentRotations(Scene):
    def construct(self):
        left_square = Square(color=BLUE, fill_opacity=0.7).shift(2 * LEFT)
        right_square = Square(color=GREEN, fill_opacity=0.7).shift(2 * RIGHT)
        self.play(
            left_square.animate.rotate(PI),
            Rotate(right_square, angle=PI),
            run_time=2
            )
        self.wait()
```

```python
%%manim -v WARNING -qm --renderer=cairo DifferentRotations
```

# Transformations

The difference between Transform and ReplacementTransform is that Transform(mob1, mob2) transforms the points (as well as other attributes like color) of mob1 into the points/attributes of mob2.

ReplacementTransform(mob1, mob2) on the other hand literally replaces mob1 on the scene with mob2.

```python
class TwoTransforms(Scene):
    def transform(self):
        a = Circle()
        b = Square()
        c = Triangle()
        self.play(Transform(a, b))
        self.play(Transform(a, c))
        self.play(FadeOut(a))

    def replacement_transform(self):
        a = Circle()
        b = Square()
        c = Triangle()
        self.play(ReplacementTransform(a, b))
        self.play(ReplacementTransform(b, c))
        self.play(FadeOut(c))

    def construct(self):
        self.transform()
        self.wait(0.5)  # wait for 0.5 seconds
        self.replacement_transform()
        self.wait(0.5)  # wait for 0.5 seconds
```

```python
%%manim -v WARNING -qm --renderer=cairo TwoTransforms
```

# Composit Structures

```python
class Slider(VGroup):
    def __init__(struct, start=LEFT * 3, end=RIGHT * 3, color=WHITE, dot_color=RED, **kwargs):
        super().__init__(**kwargs)
        struct.line_start = start
        struct.line_end = end
        struct.slider_line = Line(struct.line_start, struct.line_end, color=color)
        struct.slider_dot = Dot(color=dot_color).scale(2)
        struct.add(struct.slider_line, struct.slider_dot)
        struct.Set(0) # Initialize dot at the start

    def Set(self, proportion):
        """
        Moves the slider dot to a position along the line based on a proportion (0 to 1).
        """
        start = self.slider_line.get_start()
        end = self.slider_line.get_end()
        new_position = (1-proportion) * start + proportion * end
        return self.slider_dot.animate.move_to(new_position)


    def Slide(self, proportion, **kwargs):
        """
        Animates the slider dot to a position along the line based on a proportion (0 to 1).
        """
        start = self.slider_line.get_start()
        end = self.slider_line.get_end()
        new_position = (1-proportion) * start + proportion * end
        return self.slider_dot.animate(**kwargs).move_to(new_position)
```

```python

class SliderExample(Scene):
    def construct(self):
        # Create a horizontal slider
        horizontal_slider = Slider(start=LEFT * 4, end=RIGHT * 4)
        self.play(Create(horizontal_slider))
        self.play(horizontal_slider.Slide(0.175, run_time=2))
        self.play(horizontal_slider.Set(0.75))
        self.play(horizontal_slider.Set(0.2))
        self.play(horizontal_slider.Slide(0.72, run_time=2))
        self.wait()

        # Create a vertical slider
        vertical_slider = Slider(start=DOWN * 3, end=UP * 3, color=BLUE, dot_color=YELLOW)
        self.play(Create(vertical_slider))
        self.play(vertical_slider.Slide(0.5, run_time=2))
        self.play(vertical_slider.Slide(1, run_time=2))
        self.play(vertical_slider.Set(0.5))
        self.play(vertical_slider.Set(1))
        self.wait()

        # Create another horizontal slider and position it
        another_slider = Slider(start=LEFT * 2, end=RIGHT * 2, color=GREEN, dot_color=PURPLE)
        another_slider.shift(DOWN * 2)
        self.play(Create(another_slider))
        self.play(another_slider.Set(0.1))
        self.play(another_slider.Slide(0.9, run_time=2))
        self.play(another_slider.Slide(0.5, run_time=2))
        self.wait()
```

```python
%%manim -v WARNING -qm --renderer=cairo SliderExample
```
